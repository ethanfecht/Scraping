```python
import requests
```

```python
user_input = input("Please enter a full Wikipedia URL: ")
```

```python
pdf_title = input("Please enter a name for your file: ")
pdf_title = pdf_title.replace(' ', '')
```

```python

website_url = requests.get(user_input).text
```

```python
print(pdf_title)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python

my_table = soup.find('table',{'class':'wikitable sortable'})
```

```python
print(my_table)
```

```python
links = my_table.findAll("a")
```

```python
print(links)
```

```python
urls = []


for link in links:
    url = link.get('href')
    
    urls.append("https://en.wikipedia.org/" +url)
   

```

```python
print(urls)
```

```python
image_urls = []

for url in urls:
    if ('.jpg' or '.png') in url:
        print(url)
        image_urls.append(url)
        
 
```

```python
print(image_urls)
```

```python
import pandas as pd
df=pd.DataFrame(image_urls,columns=['URLs'])

df
```

```python
df.to_csv(r'Images'+pdf_title+'.csv', index = False)
```
